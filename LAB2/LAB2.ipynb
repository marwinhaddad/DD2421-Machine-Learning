3. The non-linear kernels have parameters; explore how they influence
the decision boundary. Reason about this in terms of the biasvariance trade-off.

```python
data = Data(sepx={'A': 0.4, 'B': -0.3}, sepy={'A': 0.5, 'B': 0.5})  # sepx={'A': 0.4, 'B': -0.3}. sepy={'A': 0.5, 'B': 0.5}
```

```python
data.plot()
```

```python
print('Generating solutions for polynomial kernel with different p values')
for p in [1, 2, 4, 8, 10, 20]:
    print('p:', p)
    svm = SVM('polynomial', data, degree=p, slack=False)
    svm.optimize()
    svm.plot_boundary()

print()
print('Generating solutions for RBF kernel with different sigma values')
svm.set_kernel('RBF')
for sigma in [0.1, 0.2, 0.5, 1, 2, 4]:
    print('Sigma:', sigma)
    svm = SVM('RBF', data, sigma=sigma, slack=False)
    svm.optimize()
    svm.plot_boundary()
```

From the result we can observe that for larger values of...

- p the complexity of the model increses. As the order of the polynimial kernel increases, there is an increase in the variance, seeing as the decision boundary goes from a linear one to a more nuanced one. This increase in variance also showcases a decrease in bias as the generality of the model becomes less prominent as the order of the polynimal kernel increases.

- sigma the model becomes more and more general, as the model goes from a tightly fitting one to one that is almost linear. sigma = 0.1 shows a decision boundary that hugs the data points tightly, while sigma=4 is linear in nature.

4. Explore the role of the slack parameter C. What happens for very large/small values?

The slack parameter C allows us to introduce some missclassifikation in the model, making the model more flexible, as there might not exist a seperating hyperplane for all datasets given some kernel. We will use the same dataset that all the kernel failed to find an optimal solution for earlier.

```python
data = Data(shape='double_arc', N=200, radius=0.5, thickness=0.1, seperation=-0.25)
```

```python
data.plot()
```

```python
svm_linear = SVM('linear', data)
svm_polynomial = SVM('polynomial', data)
svm_rbf = SVM('RBF', data)
```

```python
for svm, name in zip([svm_linear, svm_polynomial, svm_rbf], ['linear', 'polynomial', 'RBF']):
    print(name) 
    for C in [0.5, 1, 10]:
        print('C:', C)
        svm.set_slack(C=C)
        svm.optimize()
        svm.plot_boundary()
```

We se from the result that introducing lower value C will in fact introduce more slack, seeing from the larger boundary margin, and introducing a higher value C will yield less slack, seeing that the margin becomes tighter as the value of C increases.

5. magine that you are given data that is not easily separable. When
should you opt for more slack rather than going for a more complex
model (kernel) and vice versa?

There are different reasons as to why one would prefere more slack/a more complex model.

MORE SLACK:
- Noise: One would prefere more slack instead of opting for a more complex model if the data is very noisy. Training a complex model on noisy data might increse the possibility of the model capturing the noise.

- Overfitting: One might prefere to opt for a more simple model while allowing for some slack instead of opting for a more complex model to reduce the risk of overfitting the model to the data.

MORE COMPLEX MODEL:
- If the data is clearly structured in a way that might indicate that a linear model might not capture its complexity it might be better to opt for a more complex model.

- If increasing slack does not increse the models preformance it might be worth looking into trying more complex models that might performe better in validation testing.


EXAMPLE:
If the data looks like it is linearly seperable, aside from some noise or a couple of outliers, it would be more reasonable to opt for some slack instead of a more complex model to retain the generality of the model.

If the data does not seem to follow any linearly seperable pattern it might be better to opt for a more complex model since it might be better at capturing the true characteristics of the data.
